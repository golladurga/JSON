```python
import json

class Employee:
    def __init__(self, name, dob, height, city, state):
        self.name = name
        self.dob = dob
        self.height = height
        self.city = city
        self.state = state

# Creating employee objects
employees = [
    Employee("John Doe", "1990-01-01", 175, "New York", "NY"),
    Employee("Jane Smith", "1985-05-10", 165, "Los Angeles", "CA"),
    Employee("Michael Johnson", "1992-12-15", 180, "Chicago", "IL"),
    Employee("Emily Davis", "1988-07-20", 160, "Houston", "TX"),
    Employee("Daniel Wilson", "1995-03-05", 170, "Miami", "FL")
]

# Writing employee information to JSON file
data = []
for employee in employees:
    employee_data = {
        "Name": employee.name,
        "DOB": employee.dob,
        "Height": employee.height,
        "City": employee.city,
        "State": employee.state
    }
    data.append(employee_data)

with open("employee.json", "w") as file:
    json.dump(data, file)

# Reading employee information from JSON file
employee_list = []
with open("employee.json", "r") as file:
    json_data = json.load(file)
    for employee_data in json_data:
        employee = Employee(employee_data["Name"], employee_data["DOB"],
                            employee_data["Height"], employee_data["City"],
                            employee_data["State"])
        employee_list.append(employee)

# Printing the list of Employee objects
for employee in employee_list:
    print("Name:", employee.name)
    print("DOB:", employee.dob)
    print("Height:", employee.height)
    print("City:", employee.city)
    print("State:", employee.state)
    print()

    
import json

# Creating the dictionary
states = {
    "Andhra Pradesh": "Hyderabad",
    "Arunachal Pradesh": "Itanagar",
    "Assam": "Dispur",
    "Bihar": "Patna",
    "Gujarat": "Gandhinagar",
    "Karnataka": "Bengaluru",
    "Maharashtra": "Mumbai"
}

# Writing the dictionary to JSON file
with open("states.json", "w") as file:
    json.dump(states, file)

print("States and capitals dictionary has been written to states.json.")



```
